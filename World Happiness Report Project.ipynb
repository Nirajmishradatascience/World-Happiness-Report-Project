# Matrices

```python
WHR.shape
```

```python
Southern_Asia = []
Sub_Saharan_Africa = []
Eastern_Asia = []
Central_and_Eastern_Europe = []
Southeastern_Asia = []
Latin_America_and_Caribbean = []
Middle_East_and_Northern_Africa = []
Australia_and_New_Zealand = []
North_America = []
Western_Europe = []

for x in WHR['Region']:
    if x == 'Southern Asia':
         Southern_Asia.append(1)
    else:
        Southern_Asia.append(0)

for x in WHR['Region']:
    if x == 'Sub-Saharan Africa':
         Sub_Saharan_Africa.append(1)
    else:
        Sub_Saharan_Africa.append(0)

for x in WHR['Region']:
    if x == 'Eastern Asia':
         Eastern_Asia.append(1)
    else:
        Eastern_Asia.append(0)

for x in WHR['Region']:
    if x == 'Central and Eastern Europe':
         Central_and_Eastern_Europe.append(1)
    else:
        Central_and_Eastern_Europe.append(0)

for x in WHR['Region']:
    if x == 'Southeastern Asia':
         Southeastern_Asia.append(1)
    else:
        Southeastern_Asia.append(0)

for x in WHR['Region']:
    if x == 'Latin America and Caribbean':
         Latin_America_and_Caribbean.append(1)
    else:
        Latin_America_and_Caribbean.append(0)

for x in WHR['Region']:
    if x == 'Middle East and Northern Africa':
         Middle_East_and_Northern_Africa.append(1)
    else:
        Middle_East_and_Northern_Africa.append(0)

for x in WHR['Region']:
    if x == 'Australia and New Zealand':
         Australia_and_New_Zealand.append(1)
    else:
        Australia_and_New_Zealand.append(0)

for x in WHR['Region']:
    if x == 'North America':
         North_America.append(1)
    else:
        North_America.append(0)

for x in WHR['Region']:
    if x == 'Western Europe':
         Western_Europe.append(1)
    else:
        Western_Europe.append(0)

Matrix = pd.DataFrame(index=WHR.index)

Matrix['Southern Asia'] = Southern_Asia
Matrix['Sub-Saharan Africa'] = Sub_Saharan_Africa
Matrix['Eastern Asia'] = Eastern_Asia
Matrix['Central and Eastern Europe'] = Central_and_Eastern_Europe
Matrix['Southeastern Asia'] = Southeastern_Asia
Matrix['Latin America and Caribbean'] = Latin_America_and_Caribbean
Matrix['Middle East and Northern Africa'] = Middle_East_and_Northern_Africa
Matrix['Australia and New Zealand'] = Australia_and_New_Zealand
Matrix['North America'] = North_America
Matrix['Western Europe'] = Western_Europe
```
